### Excel Manipulation

```python
df = pd.read_excel("stock_data.xlsx", "Sheet1")
df
```

```python
#As we know the CEO of walmart which is not available in CSV 
#we can add data convert data in excel as

#define function

def convert_people_cell(cell):
    if cell=="n.a.":
        return 'sam walton'
    return cell

def convert_eps_cell(eps):
    if eps=="not available":
        return None
    return eps

df = pd.read_excel('stock_data.xlsx',"Sheet1", converters={
    'people':convert_people_cell,
    'eps': convert_eps_cell})
df
```

```python
df.to_excel("new.xlsx",sheet_name="stocks",index=False)
```

```python
# Writing 2 data in single excel file
# Lets 1st Create 2 Dataframes

df_stocks = pd.DataFrame({
    'tickers':['Google','Walmart','Microsoft'],
    'price':[845,65,64],
    'pe':[30.37,14.26,30.97],
    'eps':[27.82,4.61,2.12]
})

df_weather = pd.DataFrame({
    "Day":['1/1/2017', '1/2/2017', '1/3/2017'],
    "Temperature" : [32,35,28],
    "WindSpeed(mph)" : [6,7,2],
    "Event" : ['Rain', 'Sunny', 'Snow'] 
})
```

```python
with pd.ExcelWriter("stocks_weather.xlsx")as writer:
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
```

```python
#As date column is of string type, hence need to convert column into a date column.#convert date column into timestamp column


df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/5_handling_missing_data_fillna_dropna_interpolate/weather_data.csv", parse_dates=["day"])
df
```

```python
#check conversion
type(df.day[0])
```

```python
#setting date column as index column
df.set_index('day',inplace=True)
df
```

```python
#Method 1 fillna
new_df = df.fillna(0)
new_df
```

```python
#since event column cannt be 0 so we can define it like

new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'No Event'
})
new_df
```

```python
#also temp/windspeed data would seem weird by simply estimating it to be 0
#forward fill = ffill, simpl carry forward previous data
#backwardfill = simply copies 1step ahead data in the column
new_df = df.fillna(method='ffill')
```

```python
#axis attribute let us choose either to apply method rowwise or columnwise

#new_df = df.fillna(method='bfill', axis='columns')
#new_df
```

```python
#limiting copying data upto 1 row/column only

new_df=df.fillna(method='ffill', limit=1)
new_df
```

```python
new_df=df.interpolate()
new_df
```

```python
new_df=df.interpolate(method="time")
new_df
```

```python
#dropping na_values

new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how="all")
new_df
```

```python
new_df = df.dropna(thresh=1)
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```

```python
#Inserting Missing Dates
dt=pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
df=df.reindex(idx)
df
```

```python
# Replace method : Handle missing data
df=pd.read_csv('weather_data.csv')
df
```

```python
new_df = df.replace(-99999,np.nan)
new_df
```

```python
#replacing 2 special value with NaN
new_df = df.replace([-99999,-88888],np.nan)
new_df
```

```python
# Replacing certain values from specific columns.

new_df=df.replace({
    'temperature': -99999,
    'windspeed': [-99999,-88888],
    'event':['No Event',0],
},np.NaN)
new_df
```

```python
#Mapping values to replace with required data
new_df=df.replace({
    -99999:np.NaN,
    'No Event':'Sunny',
})
new_df
```
